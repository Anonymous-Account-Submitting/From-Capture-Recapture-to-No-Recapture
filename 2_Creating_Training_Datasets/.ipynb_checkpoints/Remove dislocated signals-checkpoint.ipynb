# Thus, these signals will be removed from the dataset.

```python
def removeOffsetSignals(dataset, info=False):
    new_Dataset = []
    count = 0
    removed = []
    Peaks = MapTool.getPeaksDataset(dataset, pad=7)
    for s in range(len(dataset)):
        if Peaks[s][0] < Peaks[s][1]:
            new_Dataset.append(dataset[s])
        else:
            removed.append(s)
            count = count +1
    print("Number of removed instances: ", count)
    if info:
        print("Removed locations: ", removed)
    print("Number of good signals: ", len(new_Dataset))
    
    return np.array(new_Dataset, dtype=object)
```
